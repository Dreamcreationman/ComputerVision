```python
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms, datasets
import matplotlib.pyplot as plt
```

```python
if torch.cuda.is_available():
    torch.set_default_tensor_type('torch.cuda.FloatTensor')
```

```python
class AlexNet(nn.Module):
    def __init__(self):
        super(AlexNet, self).__init__()
        self.c1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=96, kernel_size=11, stride=4, padding=2),
            nn.ReLU(),
            nn.LocalResponseNorm(k=2, alpha=10e-4, beta=0.75, size=5),
            nn.MaxPool2d(kernel_size=3, stride=2)
        )
        self.c2 = nn.Sequential(
            nn.Conv2d(in_channels=96, out_channels=256, kernel_size=5, stride=1, padding=2, groups=2),
            nn.ReLU(),
            nn.LocalResponseNorm(k=2, alpha=10e-4, beta=0.75, size=5),
            nn.MaxPool2d(kernel_size=3, stride=2)
        )
        self.c3 = nn.Sequential(
            nn.Conv2d(in_channels=256, out_channels=384, kernel_size=3, stride=1, padding=1, groups=1),
            nn.ReLU()
        )
        self.c4 = nn.Sequential(
            nn.Conv2d(in_channels=384, out_channels=384, kernel_size=3, stride=1, padding=1, groups=2),
            nn.ReLU(),
        )
        self.c5 = nn.Sequential(
            nn.Conv2d(in_channels=384, out_channels=256, kernel_size=3, stride=1, padding=1, groups=2),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=3, stride=2)
        )
        self.fc6 = nn.Sequential(
            nn.Linear(in_features=6*6*256, out_features=4096),
            nn.ReLU(),
            nn.Dropout(p=0.5)
        )
        self.fc7 = nn.Sequential(
            nn.Linear(in_features=4096, out_features=4096),
            nn.ReLU(),
            nn.Dropout(p=0.5)
        )
        self.fc8 = nn.Linear(in_features=4096, out_features=10)
        
    def forward(self, x):
        x = self.c5(self.c4(self.c3(self.c2(self.c1(x)))))
        x = x.view(-1, 6*6*256)
        x = self.fc8(self.fc7(self.fc6(x)))
        
        return x
```

```python
def get_parameter_number(net):
    total_num = sum(p.numel() for p in net.parameters())
    trainable_num = sum(p.numel() for p in net.parameters() if p.requires_grad)
    return {'Total': total_num, 'Trainable': trainable_num}
```

```python
net = AlexNet()
get_parameter_number(net)
```

```python
BATCH_SIZE = 64
EPOCH = 10
LEARNING_RATE = 0.01
MOMENTUM = 0.9
WEIGHT_DECAY = 0.0005

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY, momentum=MOMENTUM)
device
```

```python
transform = transforms.Compose([
    transforms.Pad(fill=0, padding=96),
    transforms.ToTensor()
])
data_train = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
train_dataset = torch.utils.data.DataLoader(data_train, batch_size=BATCH_SIZE, shuffle=True)
data_test = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
test_dataset = torch.utils.data.DataLoader(data_test, batch_size=BATCH_SIZE, shuffle=True)
```

```python
loss_data = []
acc = []
for epoch in range(EPOCH):
    for i,data in enumerate(train_dataset):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = net(inputs)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        if i % 100 == 0:
            loss_data.append(loss.item())
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(epoch, i * len(inputs), len(train_dataset.dataset), 100. * i / len(train_dataset), loss.item()))
    with torch.no_grad():
        correct = 0 
        total = 0
        for data in test_dataset:
            test_inputs, test_labels = data
            test_inputs, test_labels = test_inputs.to(device), test_labels.to(device)
            out = net(test_inputs)
            
            _, predicted = torch.max(out.data, 1)
            total += len(test_labels)
            correct += (predicted == test_labels).sum().item()
        acc.append(correct / total)
        print("第 %d 个epoch的识别率为：%d %%" % (epoch + 1, (100 * correct / total)))
torch.save(net, './model/alexnet.pkl')
```

```python
print(acc, loss_data)
```

```python
plt.subplot(2, 1, 1)
plt.plot(range(0, len(acc)), acc, 'o-')
plt.title('Test accuracy vs. epoches')
plt.ylabel('Test accuracy')
plt.subplot(2, 1, 2)
plt.plot(range(0, len(loss_data)), loss_data, '.-')
plt.xlabel('Test loss vs. batches')
plt.ylabel('Test loss')
plt.show()
```
